# K-Means Clustering Template


## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import random as rd
```

## Importing the dataset 
- customizable to different dataset
- change X matrix according to the features you want to use to group clusters

```
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the elbow method to find the optimal number of clusters
- change variable k depending on the maximum number of clusters you want to go up to


```
from sklearn.cluster import KMeans
k = 25
wcss = []
for i in range(1, k+1):
    kmeans = KMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1, k+1), wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()
```

## Training the K-Means model on the dataset
- manually change n variable according to the most optimal cluster size determined
    - defaulted to 5

```
n = 5
kmeans = KMeans(n_clusters = n, init = 'k-means++')
y_kmeans = kmeans.fit_predict(X)
```

## Visualising the clusters

```
for i in range(1, n+1):
    plt.scatter(X[y_kmeans == i-1, 0], X[y_kmeans == i-1, 1], s = 100, c = [[rd.uniform(0.1, 0.9), rd.uniform(0.1, 0.9), rd.uniform(0.1, 0.9)]], label = f'Cluster {i}')
    
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], 150, c = 'black', label = 'Centroids')
plt.title('Clusters Distribution')
plt.xlabel("x")
plt.ylabel('y')
plt.legend()
plt.show()
```
